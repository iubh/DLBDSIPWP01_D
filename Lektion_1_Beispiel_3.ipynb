# Kantendetektion auf einem Farbbild
Zur Ausführung dieses Beispiels wird neben den im Skript vorgestellten Bibliotheken auch OpenCV benötigt. Diese kannst du in einem Terminal (z.B. Anaconda Prompt) wie folgt installieren:
`pip install opencv-python`

Verwende ein eigenes Bild in einem Standardformat, dessen Dateinamen du in der entsprechenden Zeile unten eingibst.

```python
# Import der notwendigen Bibliotheken
import cv2
import matplotlib.pyplot as plt

# Einlesen des Bildes, gib hier den Pfad zu deinem eigenen Bild ein
# Es wird typischerweise im Format BGR eingelesen (3 Farbwerte pro Pixel)
image = cv2.imread("") # Gib hier den Pfad zu deinem Bild ein. Speichere dieses am besten in demselben Ordner wie dieses Notebook

# Konvertierung in eine Grauwertdarstellung (1 Helligkeitswert pro Pixel)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Anwendung des Canny Kantendetektors
# Mehr Infos findest du hier: https://docs.opencv.org/3.4/da/d22/tutorial_py_canny.html
edges = cv2.Canny(gray_image, threshold1=100, threshold2=200)

# Erzeugung einer Figure
plt.figure()
# Zwei Plots in einer Figure
plt.subplot(1, 2, 1)
plt.title("Originalbild")
# Anzeigen des Bildes
# Für die korrekte Anzeige muss BGR nach RGB umgewandelt werden
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
# Ausschalten der Achsenbeschriftungen
plt.axis("off")
# Fokus auf zweiten Subplot
plt.subplot(1, 2, 2)
plt.title("Canny Edges")
# Anzeigen der Kanten (als Grauwerte)
plt.imshow(edges, cmap="gray")
plt.axis("off")
# Anzeigen der Figure
plt.show()
```
