# Das ist eine Kosinus-Funktion

```python
# importiere notwendige Funktionalitäten
from matplotlib import pyplot
import numpy
# x-Werte generieren
x = numpy.arange(-4,4.1,0.1)
# Funktionswerte berechnen
y = numpy.cos(x)
# Funktionswerte darstellen
pyplot.plot(x, y, "-")
pyplot.title("Das ist eine Kosinus-Funktion")
pyplot.xlabel("x")
pyplot.ylabel("y")
pyplot.show()
```
