# Klassifikation mit nicht linear und linear trennbaren Daten

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.linear_model import LogisticRegression

# Beispieldaten erzeugen
X, y = make_moons(n_samples=200, noise=0.2, random_state=42)

# Klassifikator trainieren
model = LogisticRegression()
model.fit(X, y)

# Gitter für Entscheidungsfläche
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300),
                     np.linspace(y_min, y_max, 300))

Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot
plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor="k")
plt.title("Beispiel einer binären Klassifikation")
plt.xlabel("Merkmal 1")
plt.ylabel("Merkmal 2")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression

# Lineare Daten erzeugen
X, y = make_classification(n_samples=200, n_features=2, n_informative=2, n_redundant=0, 
                           n_clusters_per_class=1, class_sep=2.0, random_state=42)

# Klassifikator trainieren
model = LogisticRegression()
model.fit(X, y)

# Gitter für Entscheidungsfläche
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300),
                     np.linspace(y_min, y_max, 300))

Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot
plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor="k")
#plt.title("Beispiel einer binären Klassifikation")
plt.xlabel("Merkmal 1")
plt.ylabel("Merkmal 2")

# Speichern der Abbildung als PNG
plt.savefig("Beispiel_Klassifikation.png", format='png', dpi=300, bbox_inches='tight')
plt.show()
```

# Beispiel lineare Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Beispieldaten erzeugen
np.random.seed(0)
X = 2 * np.random.rand(50, 1)
y = 3 * X[:, 0] + 2 + 0.3*np.random.randn(50)

# Regressionsmodell trainieren
model = LinearRegression()
model.fit(X, y)

# Vorhersagen
X_test = np.linspace(0, 2, 100).reshape(-1, 1)
y_pred = model.predict(X_test)

# Plot
plt.scatter(X, y)
plt.plot(X_test, y_pred)
plt.title("Beispiel einer Regression")
plt.xlabel("Merkmal")
plt.ylabel("Zielwert")

# Speichern der Abbildung als PNG
plt.savefig("Beispiel_Regression.png", format='png', dpi=300, bbox_inches='tight')
plt.show()
```

# Beispiel Clustering

```python
#import os
#os.environ['OMP_NUM_THREADS'] = '1'

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Beispieldaten erzeugen mit größerer Streuung innerhalb der Cluster und Zentren in einem angemessenen Abstand
X, _ = make_blobs(n_samples=200, centers=3, cluster_std=1.5, center_box=(-5.0, 5.0), random_state=42)

# Clustering-Modell
model = KMeans(n_clusters=3, random_state=42)
labels = model.fit_predict(X)

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, edgecolor="k")
plt.scatter(model.cluster_centers_[:, 0],
            model.cluster_centers_[:, 1],
            marker="x", s=200, color="red", label="Cluster Zentroiden")
plt.title("Beispiel einer Clustering-Verteilung")
plt.xlabel("Merkmal 1")
plt.ylabel("Merkmal 2")
plt.legend()

# Speichern der Abbildung als PNG
plt.savefig("Beispiel_Clustering.png", format='png', dpi=300, bbox_inches='tight')
plt.show()
```
